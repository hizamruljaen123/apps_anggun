```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_decision_tree():
    # Membuat directed graph
    G = nx.DiGraph()
    
    # Menambahkan edges dengan struktur lengkap berdasarkan gambar
    G.add_edges_from([
        ("Status Prestasi", "MTK"),
        ("MTK", "Baik"),
        ("Baik", "GEO"),
        ("GEO", "Lain"),
        ("Lain", "BIN"),
        ("BIN", "Baik (BIN)"),  # Menambahkan kategori setelah BIN
        ("BIN", "Cukup (BIN)"), # Menambahkan kategori setelah BIN
        ("MTK", "Cukup"),
        ("Cukup", "BIG"),
        ("BIG", "Baik (BIG)"),  # Menambahkan kategori setelah BIG
        ("BIG", "Cukup (BIG)"), # Menambahkan kategori setelah BIG
        ("MTK", "Sangat Buruk"),
        ("Sangat Buruk", "Eko"),
        ("Eko", "Cukup (Eko)"),  # Menambahkan kategori setelah Eko
        ("Eko", "Sangat Buruk (Eko)"), # Menambahkan kategori setelah Eko
        ("MTK", "Tidak Tersedia")
    ])
    
    # Menentukan posisi node secara manual untuk layout yang rapi dan independen
    pos = {
        "Status Prestasi": (0, 4),
        "MTK": (0, 3),
        "Baik": (-2, 2),
        "GEO": (-3, 1),
        "Lain": (-4, 0),
        "BIN": (-5, -1),
        "Baik (BIN)": (-6, -2),
        "Cukup (BIN)": (-4, -2),
        "Cukup": (0, 2),
        "BIG": (-1, 1),
        "Baik (BIG)": (-2, 0),
        "Cukup (BIG)": (0, 0),
        "Sangat Buruk": (2, 2),
        "Eko": (3, 1),
        "Cukup (Eko)": (2, 0),
        "Sangat Buruk (Eko)": (4, 0),
        "Tidak Tersedia": (4, 2)
    }
    
    # Menggambar grafik
    plt.figure(figsize=(14, 8))
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", 
            font_size=10, font_weight="bold", edge_color="gray", arrows=True)
    
    # Menambahkan judul
    plt.title("Diagram Pohon Keputusan", fontsize=14, pad=10)
    
    # Menghilangkan sumbu
    plt.axis("off")
    
    # Menampilkan grafik
    plt.show()

# Memanggil fungsi untuk menggambar pohon keputusan
draw_decision_tree()
```
